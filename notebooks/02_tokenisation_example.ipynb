```python
from transformers import DistilBertTokenizer

tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')

sample = "i dont think im getting my baby them white kicks"

# Tokenise a single string
encoded = tokenizer(
    sample,
    max_length=128,
    padding='max_length',
    truncation=True,
    return_tensors='pt'   # return PyTorch tensors
)

print("Input IDs:")
print(encoded['input_ids'])
print(f"\nShape: {encoded['input_ids'].shape}")

print("\nAttention Mask:")
print(encoded['attention_mask'])

print("\nTokens (human readable):")
print(tokenizer.convert_ids_to_tokens(encoded['input_ids'][0]))
```

```python
special_sample = "<user> you are a <censored> go back to your country <number>"

encoded_special = tokenizer(
    special_sample,
    max_length=128,
    padding='max_length',
    truncation=True,
    return_tensors='pt'
)

print("Tokens (human readable):")
print(tokenizer.convert_ids_to_tokens(encoded_special['input_ids'][0])[:20])
```

The above tokenisation splits up special tokens in the data such as \<user\>. The below code block fixes this by registering them as special tokens. They will then never be split up.

```python
from transformers import DistilBertTokenizer

tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')

# All special tokens from EDA
domain_special_tokens = [
    '<user>', '<number>', '<percent>', '<censored>', '<date>',
    '<money>', '<time>', '<happy>', '<sad>', '<wink>', '<laugh>',
    '<url>', '<phone>', '<email>', '<annoyed>', '<will>',
    '<tong>', '<surprise>', '<kiss>', '<angel>'
]

# Register them
tokenizer.add_tokens(domain_special_tokens)

# Test it
special_sample = "<user> you are a <censored> go back to your country <number>"
encoded = tokenizer(
    special_sample,
    max_length=128,
    padding='max_length',
    truncation=True,
    return_tensors='pt'
)

print("Tokens after registration:")
print(tokenizer.convert_ids_to_tokens(encoded['input_ids'][0])[:15])
print(f"\nVocabulary size before: 30,522")
print(f"Vocabulary size after:  {len(tokenizer)}")
```
